## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Import Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Q9: Draw barplot represents survived people based on sex

```python
plt.figure(figsize=(6,3))
cols = ['blue','orange']


sns.barplot(x='sex', y='survived',palette=cols, data=df); # it shows the average
```

# Q10: Draw catplot represents survived people based on embarked

```python
cols = ['blue','orange']
sns.catplot(
    data=df, # Data
    kind="count", # Chart to use
    x="survived",palette=cols, # X axis
    col='embarked',
    alpha=1,
);
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
plt.figure(figsize=(15,8))

sns.boxplot(
    x="sex", 
    y="age",
    hue="pclass",
    data=df);
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
plt.figure(figsize=(10,3))

columns_of_interest = ['sibsp', 'parch', 'age', 'fare', 'survived']
df_selected = df[columns_of_interest]

correlation_matrix = df_selected.corr()

# Generating the heatmap
plt.figure(figsize=(4, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-0.2, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
sns.catplot(x="sibsp", y="survived", data=df, kind="bar")
plt.title("the relation between sibsp and survived")
plt.xlabel("Number of siblings/spouses abroad the titanic")
plt.ylabel("survived")
plt.show()
df
```
